# Data Cleaning for testing outside main data set

## Imports

```python
import numpy as np
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
import math
```

### Load and concat the data for cover songs

```python
df_covers = pd.concat([pd.read_csv('cover_mix.csv'), pd.read_csv('pop_covers.csv'), pd.read_csv('chill_covers.csv'), pd.read_csv('covers_of_popular_songs__200_hits.csv')])
df_covers
```

### Iding the columns that are relevant

```python
cols = ['Spotify ID','Artist Name(s)','Track Name','Acousticness','Danceability', 'Duration (ms)', 'Energy', 'Instrumentalness', 'Key', 'Liveness', 'Loudness', 'Mode', 'Speechiness','Tempo', 'Time Signature', 'Valence']
```

### Filtering the data frame on those columns

```python
df_covers = df_covers[cols]
```

### Doing the same for the cover songs to this playlist of musicals

```python
df_musicals = pd.read_csv('all_broadway_musicals.csv', on_bad_lines='skip')[cols]
df_musicals
```

### Renaming the columns for the models

```python
df_covers.rename(columns={"Spotify ID": "id", "Artist Name(s)": "Artist(s)", "Track Name" : "Title", "Tempo" : "BMP", "Time Signature": "Time_Signature"}, inplace=True)
df_musicals.rename(columns={"Spotify ID": "id", "Artist Name(s)": "Artist(s)", "Track Name" : "Title", "Tempo" : "BMP", "Time Signature": "Time_Signature"}, inplace=True)
```

### Export data frames as CSVs

```python
df_covers.to_csv('Clean_Covers.csv')
df_musicals.to_csv('Clean_Muse.csv')
```
