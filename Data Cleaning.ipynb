### Imports

```python
import numpy as np
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
import math
import requests
```

### Loading Data

```python
df_db = pd.read_csv('top_10000_1960-now.csv')
df_db
```

```python
df_db.info()
```

### Getting just raw ids

```python
df_db['Track URI'] = df_db['Track URI'].map(lambda x: x[14:])
df_db.drop(['Track Preview URL','Album Genres'], axis=1, inplace=True)
df_db.dropna(inplace=True)
```

```python
df_db.info()
```

### Load user's liked playlist

### More Loading Data

```python
df_giga = pd.read_csv('tracks_features.csv')
df_giga.info()
```

### Fixing the artists column

```python
df_giga['artists'] = df_giga['artists'].map(lambda x: x.strip('[]').split(',')[0] if len(x.strip('[]').split(',')) == 1 else x.strip('[]').split(','))
df_giga['artists'] = df_giga['artists'].map(lambda x: [i.strip("'") for i in x]  if type(x) is list else x.strip("'"))
df_giga['artists']
```

###  Combinding data 

```python
df = pd.DataFrame()
```

```python
df['id'] = list(df_db['Track URI']) + list(df_giga['id'])
df['Title'] = list(df_db['Track Name']) +  list(df_giga['name']) 
df['Artist(s)'] = list(df_db['Artist Name(s)']) + list(df_giga['artists'])
df['Acousticness'] = list(df_db['Acousticness'])  + list(df_giga['acousticness'])
df['Danceability'] = list(df_db['Danceability'])  + list(df_giga['danceability'])
df['Duration (ms)'] = list(df_db['Track Duration (ms)']) + list(df_giga['duration_ms'])
df['Energy'] = list(df_db['Energy']) + list(df_giga['energy'])
df['Instrumentalness'] = list(df_db['Instrumentalness']) + list(df_giga['instrumentalness'])
df['Key'] = list(df_db['Key'])  + list(df_giga['key'])
df['Liveness'] = list(df_db['Liveness']) + list(df_giga['liveness'])
df['Loudness'] = list(df_db['Loudness']) + list(df_giga['loudness'])
df['Mode'] = list(df_db['Mode']) + list(df_giga['mode'])
df['Speechiness'] =  list(df_db['Speechiness']) + list(df_giga['speechiness'])
df['BPM'] = list(df_db['Tempo']) + list(df_giga['tempo'])
df['Time_Signature'] = list(df_db['Time Signature']) + list(df_giga['time_signature'])
df['Valence'] = list(df_db['Valence']) + list(df_giga['valence'])
df
```

### Dropping nulls

```python
df.dropna(inplace=True)
df.drop_duplicates(subset='id', inplace=True)
```

### Export CSV

```python
df.to_csv('Total DB of Songs.csv')
```
